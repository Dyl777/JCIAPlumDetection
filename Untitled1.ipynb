```python
import os
import hashlib
from PIL import Image
import shutil
import pandas as pd
import numpy as np
import cv2
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import confusion_matrix, classification_report
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms
import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim.swa_utils import AveragedModel, SWALR
from torch.optim.lr_scheduler import CosineAnnealingLR
import albumentations as A
from albumentations.pytorch import ToTensorV2
from collections import defaultdict
import kagglehub
import matplotlib.pyplot as plt
import seaborn as sns
import torch.nn.functional as F

# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Create output directory in Google Drive
output_dir = '/content/drive/MyDrive/Plum_Model_Plots'
os.makedirs(output_dir, exist_ok=True)
print(f"Output directory created at: {output_dir}")
```

```python
def calculate_image_hash(image_path):
    try:
        with Image.open(image_path) as img:
            img = img.convert('RGB')
            img_bytes = img.tobytes()
            return hashlib.md5(img_bytes).hexdigest()
    except Exception:
        return None

def remove_duplicate_images(image_dir):
    hash_to_files = defaultdict(list)
    class_counts_before = defaultdict(int)
    class_counts_after = defaultdict(int)
    class_names = ['bruised', 'cracked', 'rotten', 'spotted', 'unaffected', 'unripe']

    for root, _, files in os.walk(image_dir):
        class_name = os.path.basename(root)
        if class_name in class_names:
            for file in files:
                if file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
                    file_path = os.path.join(root, file)
                    image_hash = calculate_image_hash(file_path)
                    if image_hash:
                        hash_to_files[image_hash].append((file_path, class_name))
                        class_counts_before[class_name] += 1

    duplicates_found = 0
    kept_files = []
    for image_hash, file_info in hash_to_files.items():
        if len(file_info) > 1:
            duplicates_found += len(file_info) - 1
            kept_files.append(file_info[0][0])
            class_counts_after[file_info[0][1]] += 1
            for file_path, _ in file_info[1:]:
                try:
                    os.remove(file_path)
                except Exception:
                    pass
        else:
            kept_files.append(file_info[0][0])
            class_counts_after[file_info[0][1]] += 1

    duplicate_proportions = {}
    for class_name in class_names:
        before = class_counts_before.get(class_name, 0)
        after = class_counts_after.get(class_name, 0)
        proportion = (before - after) / before * 100 if before > 0 else 0
        duplicate_proportions[class_name] = proportion

    print(f"Total duplicates removed: {duplicates_found}")
    print(f"Images retained: {len(kept_files)}")
    print("Sample Cleaned Data (images per class):")
    for class_name in class_names:
        print(f"{class_name}: {class_counts_after.get(class_name, 0)} images")

    # Plot duplicate proportions
    plt.figure(figsize=(10, 6))
    plt.bar(duplicate_proportions.keys(), duplicate_proportions.values())
    plt.xlabel('Class')
    plt.ylabel('Percentage of Duplicates Removed (%)')
    plt.title('Proportion of Duplicated Data per Class')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, 'duplicate_proportion.png'))
    plt.show()

    return duplicates_found, kept_files, class_counts_before, class_counts_after
```

```python
def load_dataset(image_dir):
    images = []
    labels = []
    class_names = ['bruised', 'cracked', 'rotten', 'spotted', 'unaffected', 'unripe']
    for idx, class_name in enumerate(class_names):
        class_dir = os.path.join(image_dir, class_name)
        if not os.path.exists(class_dir):
            print(f"Directory {class_dir} does not exist")
            continue
        for img_file in os.listdir(class_dir):
            if img_file.lower().endswith(('.png', '.jpg', '.jpeg', '.JPG')):
                img_path = os.path.join(class_dir, img_file)
                try:
                    with Image.open(img_path) as img:
                        img.verify()
                    images.append(img_path)
                    labels.append(idx)
                except Exception:
                    print(f"Invalid image: {img_path}")
    return images, labels
```

```python
def augment_image(image_path, augmentations):
    try:
        image = cv2.imread(image_path)
        if image is None:
            raise ValueError(f"Failed to load image: {image_path}")
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        augmented = augmentations(image=image)
        return augmented['image']
    except Exception:
        return None

def balance_classes(images, labels):
    if not images or not labels:
        print("Cannot balance classes: empty dataset")
        return [], []
    class_counts = pd.Series(labels).value_counts()
    print("Class counts before balancing:")
    print(class_counts.rename(index=dict(enumerate(['bruised', 'cracked', 'rotten', 'spotted', 'unaffected', 'unripe']))))

    # Plot class distribution before balancing
    plt.figure(figsize=(10, 6))
    class_counts.plot(kind='bar')
    plt.xlabel('Class')
    plt.ylabel('Number of Images')
    plt.title('Class Distribution Before Balancing')
    plt.xticks(ticks=range(len(class_counts)), labels=['bruised', 'cracked', 'rotten', 'spotted', 'unaffected', 'unripe'], rotation=45)
    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, 'class_distribution_before.png'))
    plt.show()

    max_count = class_counts.max()
    balanced_images = []
    balanced_labels = []
    augmentations = A.Compose([
        A.HorizontalFlip(p=0.5),
        A.Rotate(limit=30, p=0.3),
        A.RandomCrop(height=80, width=80, p=0.2),
        A.ColorJitter(brightness=0.1, contrast=0.1, saturation=0.1, hue=0.1, p=0.2),
        A.CoarseDropout(max_holes=8, max_height=8, max_width=8, p=0.2),  # Cutout
    ])

    for cls in range(6):
        cls_images = [img for img, lbl in zip(images, labels) if lbl == cls]
        num_to_add = max_count - len(cls_images)
        if num_to_add > 0:
            additional_images = []
            for _ in range(num_to_add):
                img_path = np.random.choice(cls_images)
                aug_img = augment_image(img_path, augmentations)
                if aug_img is not None:
                    additional_images.append(aug_img)
            balanced_images.extend(cls_images)
            balanced_images.extend(additional_images[:num_to_add])
            balanced_labels.extend([cls] * max_count)
        else:
            balanced_images.extend(cls_images[:max_count])
            balanced_labels.extend([cls] * max_count)

    # Plot class distribution after balancing
    balanced_counts = pd.Series(balanced_labels).value_counts()
    print("Class counts after balancing:")
    print(balanced_counts.rename(index=dict(enumerate(['bruised', 'cracked', 'rotten', 'spotted', 'unaffected', 'unripe']))))
    plt.figure(figsize=(10, 6))
    balanced_counts.plot(kind='bar')
    plt.xlabel('Class')
    plt.ylabel('Number of Images')
    plt.title('Class Distribution After Balancing')
    plt.xticks(ticks=range(len(balanced_counts)), labels=['bruised', 'cracked', 'rotten', 'spotted', 'unaffected', 'unripe'], rotation=45)
    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, 'class_distribution_after.png'))
    plt.show()

    return balanced_images, balanced_labels
```

```python
def display_sample_images(images, labels, class_names, num_samples=2):
    samples_per_class = defaultdict(list)
    for img_path, label in zip(images, labels):
        if len(samples_per_class[label]) < num_samples:
            samples_per_class[label].append(img_path)
        if all(len(samples_per_class[l]) >= num_samples for l in range(len(class_names))):
            break

    plt.figure(figsize=(15, 10))
    for idx, (label, img_paths) in enumerate(samples_per_class.items()):
        for i, img_path in enumerate(img_paths):
            try:
                img = Image.open(img_path).convert('RGB')
                plt.subplot(len(class_names), num_samples, idx * num_samples + i + 1)
                plt.imshow(img)
                plt.title(f"{class_names[label]}")
                plt.axis('off')
            except Exception:
                plt.subplot(len(class_names), num_samples, idx * num_samples + i + 1)
                plt.text(0.5, 0.5, 'Invalid Image', ha='center', va='center')
                plt.title(f"{class_names[label]}")
                plt.axis('off')
    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, 'sample_images.png'))
    plt.show()
    print("Sample images saved to Google Drive")
```

```python
class PlumDataset(Dataset):
    def __init__(self, images, labels, transform=None):
        self.images = images
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        try:
            if isinstance(self.images[idx], str):
                img = Image.open(self.images[idx]).convert('RGB')
            else:
                img = Image.fromarray(self.images[idx])
            label = self.labels[idx]
            if self.transform:
                img = self.transform(img)
            return img, label
        except Exception:
            return None
```

```python
class ModifiedVGG(nn.Module):
    def __init__(self, num_classes=6):
        super(ModifiedVGG, self).__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.Conv2d(128, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(128, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.Conv2d(256, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.Conv2d(256, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(256, 512, kernel_size=3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(inplace=True),
            nn.Conv2d(512, 512, kernel_size=3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(inplace=True),
            nn.Conv2d(512, 512, kernel_size=3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(512, 512, kernel_size=3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(inplace=True),
            nn.Conv2d(512, 512, kernel_size=3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(inplace=True),
            nn.Conv2d(512, 512, kernel_size=3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.avgpool = nn.AdaptiveAvgPool2d((7, 7))
        self.classifier = nn.Sequential(
            nn.Linear(512 * 7 * 7, 4096),
            nn.ReLU(True),
            nn.Dropout(),
            nn.Linear(4096, 4096),
            nn.ReLU(True),
            nn.Dropout(),
            nn.Linear(4096, num_classes),
        )

    def forward(self, x):
        x = self.features(x)
        x = self.avgpool(x)
        x = x.view(x.size(0), -1)
        x = self.classifier(x)
        return x
```

```python
class LabelSmoothingCrossEntropy(nn.Module):
    def __init__(self, smoothing=0.1):
        super(LabelSmoothingCrossEntropy, self).__init__()
        self.smoothing = smoothing
        self.confidence = 1.0 - smoothing

    def forward(self, x, target):
        logprobs = F.log_softmax(x, dim=-1)
        nll_loss = -logprobs.gather(dim=-1, index=target.unsqueeze(1))
        nll_loss = nll_loss.squeeze(1)
        smooth_loss = -logprobs.mean(dim=-1)
        loss = self.confidence * nll_loss + self.smoothing * smooth_loss
        return loss.mean()
```

```python
def mixup_data(x, y, alpha=0.2, device='cuda'):
    if alpha > 0:
        lam = np.random.beta(alpha, alpha)
    else:
        lam = 1
    batch_size = x.size()[0]
    index = torch.randperm(batch_size).to(device)  # Ensure indices are on the same device
    mixed_x = lam * x + (1 - lam) * x[index, :]
    y_a, y_b = y, y[index]
    return mixed_x.to(device), y_a.to(device), y_b.to(device), lam

def mixup_criterion(criterion, pred, y_a, y_b, lam):
    return lam * criterion(pred, y_a) + (1 - lam) * criterion(pred, y_b)
```

```python
def train_model(model, train_loader, val_loader, test_loader, criterion, optimizer, scheduler, swa_model, swa_scheduler, epochs, device, model_path='best_model.pth'):
    best_val_loss = float('inf')
    train_losses = []
    val_losses = []
    train_accuracies = []
    val_accuracies = []
    lrs = []

    # Ensure model is on the correct device
    model.to(device)

    for epoch in range(epochs):
        model.train()
        train_loss = 0.0
        train_correct = 0
        train_total = 0
        for images, labels in train_loader:
            if images is None or labels is None:
                continue
            # Move tensors to device
            images, labels = images.to(device), labels.to(device)
            # Apply mixup
            images, labels_a, labels_b, lam = mixup_data(images, labels, alpha=0.2, device=device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = mixup_criterion(criterion, outputs, labels_a, labels_b, lam)
            if torch.isnan(loss):
                continue
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()
            train_loss += loss.item() * images.size(0)
            _, predicted = torch.max(outputs, 1)
            train_total += labels.size(0)
            train_correct += (predicted == labels_a).sum().item()  # Approximate accuracy
        scheduler.step()
        lrs.append(optimizer.param_groups[0]['lr'])

        train_loss = train_loss / train_total if train_total > 0 else float('inf')
        train_acc = 100 * train_correct / train_total if train_total > 0 else 0
        train_losses.append(train_loss)
        train_accuracies.append(train_acc)

        model.eval()
        val_loss = 0.0
        val_correct = 0
        val_total = 0
        with torch.no_grad():
            for images, labels in val_loader:
                if images is None or labels is None:
                    continue
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                loss = criterion(outputs, labels)
                if torch.isnan(loss):
                    continue
                val_loss += loss.item() * images.size(0)
                _, predicted = torch.max(outputs, 1)
                val_total += labels.size(0)
                val_correct += (predicted == labels).sum().item()

        val_loss = val_loss / val_total if val_total > 0 else float('inf')
        val_acc = 100 * val_correct / val_total if val_total > 0 else 0
        val_losses.append(val_loss)
        val_accuracies.append(val_acc)

        print(f"Epoch {epoch+1}/{epochs} | Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.2f}% | "
              f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.2f}%")

        if val_loss < best_val_loss and val_total > 0:
            best_val_loss = val_loss
            torch.save({
                'epoch': epoch,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'val_loss': val_loss,
            }, model_path)
            print(f"Saved model checkpoint to {model_path}")

        if epoch >= 10:
            swa_model.update_parameters(model)
            swa_scheduler.step()

    swa_model_path = 'swa_' + model_path
    torch.save(swa_model.state_dict(), swa_model_path)
    print(f"Saved SWA model to {swa_model_path}")

    # Plot training and validation loss
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, epochs+1), train_losses, label='Training Loss')
    plt.plot(range(1, epochs+1), val_losses, label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss')
    plt.legend()
    plt.grid(True)
    plt.savefig(os.path.join(output_dir, 'loss_curves.png'))
    plt.show()

    # Plot training and validation accuracy
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, epochs+1), train_accuracies, label='Training Accuracy')
    plt.plot(range(1, epochs+1), val_accuracies, label='Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy (%)')
    plt.title('Training and Validation Accuracy')
    plt.legend()
    plt.grid(True)
    plt.savefig(os.path.join(output_dir, 'accuracy_curves.png'))
    plt.show()

    # Plot learning rate schedule
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, epochs+1), lrs, label='Learning Rate')
    plt.xlabel('Epoch')
    plt.ylabel('Learning Rate')
    plt.title('Learning Rate Schedule')
    plt.legend()
    plt.grid(True)
    plt.savefig(os.path.join(output_dir, 'lr_schedule.png'))
    plt.show()

    # Evaluate on test set
    model.eval()
    test_loss = 0.0
    test_correct = 0
    test_total = 0
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for images, labels in test_loader:
            if images is None or labels is None:
                continue
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            if torch.isnan(loss):
                continue
            test_loss += loss.item() * images.size(0)
            _, predicted = torch.max(outputs, 1)
            test_total += labels.size(0)
            test_correct += (predicted == labels).sum().item()
            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    test_loss = test_loss / test_total if test_total > 0 else float('inf')
    test_acc = 100 * test_correct / test_total if test_total > 0 else 0
    print(f"Test Loss: {test_loss:.4f} | Test Acc: {test_acc:.2f}%")

    if all_labels and all_preds:
        cm = confusion_matrix(all_labels, all_preds)
        class_names = ['bruised', 'cracked', 'rotten', 'spotted', 'unaffected', 'unripe']
        plt.figure(figsize=(10, 8))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=class_names, yticklabels=class_names)
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title('Confusion Matrix')
        plt.savefig(os.path.join(output_dir, 'confusion_matrix.png'))
        plt.show()
        print("Confusion matrix saved to Google Drive")

        report = classification_report(all_labels, all_preds, target_names=class_names, output_dict=True)
        print("Classification Report:\n" + classification_report(all_labels, all_preds, target_names=class_names))

        # Plot per-class accuracy
        per_class_acc = cm.diagonal() / cm.sum(axis=1)
        plt.figure(figsize=(10, 6))
        plt.bar(class_names, per_class_acc)
        plt.xlabel('Class')
        plt.ylabel('Accuracy')
        plt.title('Per-Class Accuracy')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.savefig(os.path.join(output_dir, 'per_class_accuracy.png'))
        plt.show()
        print("Per-class accuracy plot saved to Google Drive")

    return train_losses, val_losses, train_accuracies, val_accuracies
```

```python
def main():
    dataset_dir = kagglehub.dataset_download("arnaudfadja/african-plums-quality-and-defect-assessment-data")
    print(f"Dataset downloaded to: {dataset_dir}")
    print(f"Contents of dataset_dir: {os.listdir(dataset_dir)}")

    image_dir = os.path.join(dataset_dir, 'african_plums_dataset', 'african_plums')
    writable_dir = '/african_plums_cleaned'
    os.makedirs(os.path.dirname(writable_dir), exist_ok=True)
    if os.path.exists(writable_dir):
        shutil.rmtree(writable_dir)
    shutil.copytree(image_dir, writable_dir)
    print(f"Copied dataset to: {writable_dir}")

    duplicates_removed, kept_files, class_counts_before, class_counts_after = remove_duplicate_images(writable_dir)
    print(f"Images after duplicate removal: {len(kept_files)}")

    images, labels = load_dataset(writable_dir)
    if not images:
        print("No images found in the dataset. Check directory structure and file extensions.")
        return
    print(f"Initial Dataset: {len(images)} images")

    class_names = ['bruised', 'cracked', 'rotten', 'spotted', 'unaffected', 'unripe']
    display_sample_images(images, labels, class_names)

    balanced_images, balanced_labels = balance_classes(images, labels)
    if not balanced_images:
        print("Failed to balance classes. Exiting.")
        return
    print(f"Balanced Dataset: {len(balanced_images)} images")

    # 5-fold cross-validation
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    fold_val_accs = []
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Using device: {device}")
    epochs = 100  # Reduced for cross-validation

    for fold, (train_val_idx, test_idx) in enumerate(kf.split(balanced_images)):
        print(f"\nFold {fold+1}/5")
        train_val_images = [balanced_images[i] for i in train_val_idx]
        train_val_labels = [balanced_labels[i] for i in train_val_idx]
        test_images = [balanced_images[i] for i in test_idx]
        test_labels = [balanced_labels[i] for i in test_idx]

        train_images, val_images, train_labels, val_labels = train_test_split(
            train_val_images, train_val_labels, test_size=0.125, stratify=train_val_labels, random_state=42
        )

        print(f"Fold {fold+1} - Train: {len(train_images)}, Val: {len(val_images)}, Test: {len(test_images)}")

        transform = transforms.Compose([
            transforms.Resize((100, 100)),  # Original ModifiedVGG input size
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
        ])

        train_dataset = PlumDataset(train_images, train_labels, transform=transform)
        val_dataset = PlumDataset(val_images, val_labels, transform=transform)
        test_dataset = PlumDataset(test_images, test_labels, transform=transform)
        train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=2)
        val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=2)
        test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False, num_workers=2)

        model = ModifiedVGG(num_classes=6).to(device)
        # Increase dropout in classifier
        model.classifier = nn.Sequential(
            nn.Linear(512 * 7 * 7, 4096),
            nn.ReLU(True),
            nn.Dropout(p=0.7),
            nn.Linear(4096, 4096),
            nn.ReLU(True),
            nn.Dropout(p=0.7),
            nn.Linear(4096, 6),
        ).to(device)  # Ensure classifier is on device
        criterion = LabelSmoothingCrossEntropy(smoothing=0.1)
        optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9, weight_decay=1e-3)
        scheduler = CosineAnnealingLR(optimizer, T_max=epochs)
        swa_model = AveragedModel(model).to(device)
        swa_scheduler = SWALR(optimizer, swa_lr=0.0005)

        train_losses, val_losses, train_accuracies, val_accuracies = train_model(
            model, train_loader, val_loader, test_loader, criterion, optimizer, scheduler,
            swa_model, swa_scheduler, epochs, device, model_path=f'best_model_fold{fold+1}.pth'
        )

        # Evaluate final validation accuracy
        model.eval()
        val_correct = 0
        val_total = 0
        with torch.no_grad():
            for images, labels in val_loader:
                if images is None or labels is None:
                    continue
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                _, predicted = torch.max(outputs, 1)
                val_total += labels.size(0)
                val_correct += (predicted == labels).sum().item()
        val_acc = 100 * val_correct / val_total if val_total > 0 else 0
        fold_val_accs.append(val_acc)
        print(f"Fold {fold+1} Validation Accuracy: {val_acc:.2f}%")

    print(f"\nAverage Validation Accuracy: {np.mean(fold_val_accs):.2f}% ± {np.std(fold_val_accs):.2f}%")

if __name__ == '__main__':
    main()
```
